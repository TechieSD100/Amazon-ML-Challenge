### Preprocessing the images

Convert the downloaded images into a format that can be processed by the model (e.g., resizing, normalizing).

```python
import os
from PIL import Image
import numpy as np

def preprocess_image(image_path, size=(224, 224)):
    img = Image.open(image_path)
    img = img.resize(size)
    img = np.array(img) / 255.0  # Normalize
    return img

# Directory path where images are stored
image_directory = '../images'

# Loop through each image in the directory and preprocess it
for image_file in os.listdir(image_directory):
    image_path = os.path.join(image_directory, image_file)
    
    # Check if it's a valid image file (this filters out directories or other non-image files)
    if os.path.isfile(image_path):
        try:
            # Preprocess each image
            processed_image = preprocess_image(image_path)
            print(f"Processed {image_file}")
        except Exception as e:
            print(f"Error processing {image_file}: {e}")
```
